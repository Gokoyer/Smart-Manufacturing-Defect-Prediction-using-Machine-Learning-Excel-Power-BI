```python
#===============================
#Section 1 — Import Libraries
#===============================
# Data processing
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Machine learning tools
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
#=========================#
#Section 2 — Load Dataset
#=========================
# Load manufacturing dataset
df = pd.read_csv("manufacturing_iot_quality_dataset.csv")

# Display first few rows
df.head()
```

```python
#============================
#Section 3 — Dataset Overview
#============================
# Show dataset shape (rows, columns)
df.shape

# Display column data types
df.info()

# Summary statistics
df.describe()
```

```python
#============================
#Section 4 — Missing Values
#============================
# Check missing values
df.isnull().sum()

If any exist:

# Replace missing values with column mean
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
#========================================
#Section 5 — Exploratory Data Analysis
#========================================
# Defect Distribution
sns.countplot(x="defect_flag", data=df)

plt.title("Defect Distribution")

plt.show()
```

```python
#==============================
#Machine Vibration vs Defects
#===============================
sns.boxplot(x="defect_flag", y="vibration_rms", data=df)

plt.title("Machine Vibration vs Defects")

plt.show()
```

```python
#========================
#Temperature vs Defects
#========================
sns.boxplot(x="defect_flag", y="ambient_temp_c", data=df)

plt.title("Temperature vs Defects")

plt.show()
```

```python
#=====================
#Correlation Heatmap
#=====================
corr = df.corr(numeric_only=True)

sns.heatmap(corr, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

```python
#===============================
#Section 6 — Feature Engineering
#=================================

#Convert categorical variables to numbers.

# Initialize encoder
le = LabelEncoder()

categorical_cols = [
    "plant",
    "line_id",
    "machine_id",
    "shift",
    "material_batch",
    "defect_type"
]

# Encode categorical columns
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
#=======================================
#Section 7 — Define Features and Target
#=======================================
# Features
X = df.drop("defect_flag", axis=1)

# Target variable
y = df["defect_flag"]
```

```python
#=============================
#Section 8 — Train Test Split
#==============================
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)
```

```python
#==========================================
#Section 9 — Train Machine Learning Model
#===========================================
#Random Forest
from sklearn.ensemble import RandomForestClassifier

# Initialize model
rf = RandomForestClassifier()

# Train model
rf.fit(X_train, y_train)

# Predict results
y_pred = rf.predict(X_test)
```

```python
#===============================
#Section 10 — Model Evaluation
#===============================
# Show classification metrics
print(classification_report(y_test, y_pred))
```

```python
#Accuracy
accuracy_score(y_test, y_pred)
```

```python
#========================
#Confusion Matrix
#========================
cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True)

plt.title("Confusion Matrix")

plt.show()
```
